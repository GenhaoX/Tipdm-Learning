# 三表拼接函数

```python
def union():
    table1=pd.read_csv('./上月订单数据.csv',encoding = 'gbk')
    table2 = pd.read_csv('./上半月订单数据.csv',encoding='gbk')
    table3 = pd.read_csv('./近7天订单数据.csv',encoding='gbk')
    tables = [table1,table2,table3]
    table = pd.concat(tables)
    return table
```
