```python
class car:
    def __init__(self,brand,model,year,speed):
        self.brand=brand
        self.model=model
        self.year=year
        self.speed=speed
    def accelerate(self,increment):
        print(f'汽车加速了{increment}')
        self.speed=self.speed+increment
        print(f'加速后速度为{self.speed}')
    def brake(self,decrement):
        self.speed=self.speed-decrement
        if self.speed<=0:
            print(f'减速了{decrement}')
            print('汽车减速到停止了')
        else:
            print(f'汽车减速了{decrement}')
            print(f'减速后速度为{self.speed}')
    def get_speed(self):
        if self.speed<=0:
            print('现在汽车停止了,速度为0')
        else:
            print(f'汽车现在速度为{self.speed}')
```

```python
b=car(1,2,4,0)
b.get_speed()
b.accelerate(10)
b.brake(8)
b.accelerate(10)
b.accelerate(20)
b.accelerate(50)
b.brake(70)
b.get_speed()
```

```python
import shutil
import os
def copy_directory(source_dir, destination_dir):
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)
        print("目标文件夹为空，正在重新创建目标文件夹")
    else:
        for file in os.listdir(source_dir):
            file_path = os.path.join(source_dir, file)
            dst_path = os.path.join(destination_dir, file)
            if os.path.isfile(file_path):
                shutil.copy2(file_path, dst_path)
            else:
                copy_directory(file_path, dst_path)
                print("存在多级文件夹，正在复制。")

                

```
