```python
class Car:
    def __init__(self,brand,model,year,speed):
        self.brand = brand
        self.model = model
        self.year = year
        self.speed = speed
    def accelerate(self,increment):
        self.speed = self.speed + increment #更新当前车速
        if self.speed >=300:#当车速等于，高于300时，车速为300
            self.speed =300
        print(f'加速了{increment}')
        print(f'加速后，车速为：{self.speed}') 
    def brake(self,decrement):
        self.speed = self.speed - decrement #更新当前车速
        if self.speed <=0:#当车速等于，低于0时，车速为0
            self.speed =0
        print(f'减速了{decrement}')
        print(f'减速后，车速为：{self.speed}')
    def get_speed(self):
        return self.speed
```

```python
import os
import shutil
def copy_directory(source_dir,destination_dir):
    if not os.path.exists(destination_dir):
        print('该文件不存在，正在创建中')
        os.makedirs(destination_dir)# 创建文件夹
        print('文件创建完成') 
    else :
        for root, _, fnames in os.walk(source_dir):
            for fname in sorted(fnames): # sorted函数把遍历的文件按文件名排序
                fpath = os.path.join(root, fname)
                shutil.copy(fpath, destination_dir) # 完成文件拷
            print('内容已复制成功！')
```
