通过input输入一个字符，并判断它的类型，然后输出它是什么类型。 可以无限次输入，直至自己输入‘退出’才会结束。 例如：
[1,24] --->print('您输入的是列表')
Csd --->print('您输入的是字符串')

```python
import string
inpt=input('请输入一个字符:')
i=1
for i in range(1,i+1):
    if inpt=='退出':
        break
    else:
        if type(eval(inpt))==list:
            print('您输入的是列表')
        if type(eval(inpt))==int or type(eval(inpt))==float or type(eval(inpt))==complex:
            print('您输入的是数字')
        if type(eval(inpt))==set:
            print('您输入的是集合')
        if type(eval(inpt))==dict:
            print('您输入的是字典')
        if type(eval(inpt))==list:
            print('您输入的是列表')
        if type(eval(inpt))==tuple:
            print('您输入的是元组')
        if type(eval(inpt))==str:
            print('您输入的是字符串')
```

定义了一个 Car 类，它有 brand、model 和 year 属性，表示汽车的品牌、型号和制造年份。还有 speed 属性表示当前车速。
类中定义三个方法：accelerate（加速）、brake（刹车）和get_speed（获取速度）。
(1)accelerate可以输入加速参数increment表示加速了多少，输出经过加速后汽车的速度
(2)brake可以输入参数decrement表示减速了多少，输出减速后的汽车速度
(3)get_speed可以获取当前汽车的速度

```python
avg_num
```

```python
class Car:
    def __init__(self,brand,model,year,speed):
        self.brand=brand
        self.model=model
        self.year=year
        self.speed=speed
    def accelerate(self,increment):
        print(f'增加的速度为{increment}')
        self.speed=self.speed+increment
        print(f'加速后的速度为{self.speed}')
    def brake(self,decrement):
        print(f'减少的速度为{decrement}')
        self.speed=self.speed-decrement
        print(f'减速后的速度为{self.speed}')
    def get_speed(self):
        return self.speed
```

```python
#实例化 
car1=Car(brand='audi',model='a8l',year='2020',speed=90)
car1.brake(24)
```

5.内容判断

a = [1,2,3] 与 b = [3,2,1]，两个列表中的内容是一样的，怎么判断出来？

```python
a = [1,2,3] 
b = [3,2,1]
a=set(a)
b=set(b)

if len(set(a.difference(b)))==0:
    print(f'a和b的差集为空，所以a、b两个列表的内容是一样的',sep='，')
```
