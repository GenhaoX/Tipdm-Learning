# # 第三题

```python
path=r"C:\Users\伍文浩\Documents\XiaoMiNet\Upupoo\Docker\config\作业\实验5 - 材料\python基础\python基础\data\num.txt"
with open(path, 'r') as f:
    num=f.read()
```

```python
import re
num1=num
numbers=[]
for i in eval(num1):
    numbers.extend(eval(i))
    
# 将列表中的字符串取出
number_detil=[]
for i in numbers:
    number_detil.append(re.findall(r'\d+', i)) # 对于数字符号取出，再拼接
    
    
avg_number = []
for i in number_detil:
    if len(i) > 0:
        avg = sum([int(x) for x in i]) / len(i) # 使用列表推导式，对字符串进行int()（
        avg_number.append(avg)
    else:
        avg_number.append(0)  # 默认值为0

avg_number
```

# #第六题

```python
import os
import shutil

def copy_directory(source_dir, destination_dir):
    # 判断目标文件夹是否存在，如果不存在则新建
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)
        print("文件夹创建成功")

    # 复制操作
    for item in os.listdir(source_dir):
        # 文件路径的拼接
        source_item = os.path.join(source_dir, item)
        destination_item = os.path.join(destination_dir, item)
        if os.path.isdir(source_item):
            # 如果是子文件夹，则递归方法来复制子文件夹
            copy_directory(source_item, destination_item)
        else:
            # 如果是文件，则使用 shutil.copy2 进行复制操作
            shutil.copy2(source_item, destination_item)
    print("复制成功")


    
    
    
source_directory = r"C:\Users\伍文浩\Desktop\data"
destination_directory = r"C:\Users\伍文浩\Desktop\date1"

copy_directory(source_directory, destination_directory)
```

```python
# 打印棋盘
def print_board(board):
    for i in range(len(board)):
        print(" | ".join(board[i]))
        if i < len(board) - 1:
            print("-" * (len(board) * 4 ))

board = [['  ','  ','  '],
         ['  ','  ','  '],
         ['  ','  ','  ']]

print_board(board)
```
